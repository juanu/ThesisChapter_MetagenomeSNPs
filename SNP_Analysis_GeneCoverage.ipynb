```python
jan_freq_data = []
aug_freq_data = []

genome_zero_count = defaultdict()

for genome in genomes_order:
    jan = []
    aug = []
    
    jan_zero = 0
    aug_zero = 0
    
    for line in open(output_folder + genome + ".genecoverage", 'r'):
        if line.strip():
            elements = line.split("\t")
            gene_id =  elements[8].split(";")[0].split("=")[1]
            jan_count,jan_freq,aug_count,aug_freq = elements[9:13]
            jan.append(float(jan_freq)*100)
            aug.append(float(aug_freq)*100)
            
            if int(jan_count) == 0:
                jan_zero += 1
            if int(aug_count) == 0:
                aug_zero += 1
    
    genome_zero_count[genome] = (jan_zero, aug_zero)
    
    jan_freq_data.append(jan)
    aug_freq_data.append(aug)
    
    
    
```

```python
#Use custom parameters
from matplotlib import rcParams
rcParams['axes.labelsize'] = 18 #18
rcParams['xtick.labelsize'] = 15 # 15
rcParams['ytick.labelsize'] = 15 #15
rcParams['legend.fontsize'] = 14 # 18
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Computer Modern Roman']
rcParams['text.usetex'] = False
rcParams['figure.figsize'] = 5.8,3.5
```

```python
fig = plt.figure(figsize=(19,8))
sns.boxplot(jan_freq_data,color="Set1", widths=.5, names=genomes_order)
plt.ylabel("% of SNPs")
fig.tight_layout()
fig.savefig("Jan_SNP_freq_boxplot.pdf")
```

```python
fig = plt.figure(figsize=(19,8))
sns.boxplot(aug_freq_data,color="Set1", widths=.5, names=genomes_order)
plt.ylabel("% of SNPs")
fig.tight_layout()
fig.savefig("Aug_SNP_freq_boxplot.pdf")
```
