```python
one_letter_code_cog = defaultdict(list)
for line in open("list_cog", 'r'):
    if line.strip():
        line = line.rstrip()
        letter, number, description = line.split("\t")
        
        for char in letter:
            if char=="[" or char=="]":
                continue
            else:
                one_letter_code_cog[number].append(char)
```

```python
import cPickle as pickle
pN_pS_Jan = pickle.load(open("pN_pS_Jan.dict"))
pN_pS_Aug = pickle.load(open("pN_pS_Aug.dict"))
```

```python
annotation_folder = "/home/ubuntu/LT_data/CompGenomics/LT_Genomes/v2/processed_genomes/annotation/"
```

```python
#Store the COG count for each genome
all_genome_cog = defaultdict(lambda:defaultdict(int))
cog_definition = defaultdict()
gene_annotation = defaultdict(lambda:defaultdict())
scaf_gene = defaultdict(list)

genome_letter_cog = defaultdict(lambda:defaultdict(int))

for genome in genomes_order:
    for line in open(annotation_folder + genome + ".txt",'r'):
        if line.strip():
            line = line.rstrip()
            elements = line.split("\t")
            
            gene_id = elements[0]
            
            
            if elements[1] == "start":
                gene_annotation[gene_id]["start"] = elements[2]
            
            if elements[1] == "end":
                gene_annotation[gene_id]["end"] = elements[2]
                
            if elements[1] == "contig":
                gene_annotation[gene_id]["scaf"] = elements[2]
                scaf_gene[elements[2]].append(gene_id)
                
            if elements[1] == "Product":
                gene_annotation[gene_id]["product"] = elements[2]
                
            if elements[1] == "COG":
                gene_annotation[gene_id]["COG"] = elements[2]
                all_genome_cog[genome][elements[2]] += 1
                cog_definition[elements[2]] = elements[3]
                
                for entry in one_letter_code_cog[elements[2]]:
                    genome_letter_cog[genome][entry] += 1
    
                
```

```python
jan_average_pn_ps_genome = defaultdict()
aug_average_pn_ps_genome = defaultdict()

average_pn_ps_cogs = defaultdict()
cog_count_positive_selection = defaultdict()

sample_cog_pos_count = defaultdict()

product_count_positive_selection = defaultdict()

product_to_print = []

cog_gene_description_positive_selection = defaultdict(list)

for genome in genomes_order:
    genome_pos_count = defaultdict()
    genome_avg_cog = defaultdict()
    genome_prod_count = defaultdict()
    
    jan_average_pn_ps_genome[genome] = sum([float(y[2]) for y in pN_pS_Jan[genome].values() if not (np.isnan(float(y[2])) or
                                                                                                    np.isinf(float(y[2])))]) \
                                       / len([y[2] for y in pN_pS_Jan[genome].values() if not (np.isnan(float(y[2])) or
                                                                                                    np.isinf(float(y[2])))])
        
    aug_average_pn_ps_genome[genome] = sum([float(y[2]) for y in pN_pS_Aug[genome].values() if not (np.isnan(float(y[2])) or
                                                                                                    np.isinf(float(y[2])))]) \
                                       / len([y[2] for y in pN_pS_Aug[genome].values() if not (np.isnan(float(y[2])) or
                                                                                                    np.isinf(float(y[2])))])
    
    #January
    jan_cog_pos_count = defaultdict(int)
    jan_avg_cog = defaultdict(list)
    
    jan_product_pos_count = defaultdict(int)
    
    for gene in pN_pS_Jan[genome]:
        dn, ds, dn_ds = pN_pS_Jan[genome][gene]
                
        if np.isnan(float(dn_ds)):
            continue
            
        if "product" in gene_annotation[gene]:
            
            product = gene_annotation[gene]["product"]
            if float(dn_ds) > 1:                
                product_to_print.append(["January",genome,gene,product,dn,ds,dn_ds])
            if np.isinf(float(dn_ds)):
                product_to_print.append(["January",genome,gene,product,dn,ds,dn_ds])

        
        if np.isinf(float(dn_ds)):
            continue
                
        if "COG" in gene_annotation[gene]:
            cog_number = gene_annotation[gene]["COG"]
        
            jan_avg_cog[cog_number].append(dn_ds)
            
            if float(dn_ds) > 1:
                jan_cog_pos_count[cog_number] += 1
                cog_letters = one_letter_code_cog[cog_number]
                
                for entry in cog_letters:
                    cog_gene_description_positive_selection[genome].append(["january",genome,gene,cog_number,cog_letter_desc[entry],str(dn_ds)])
                
            #elif not np.isfinite(float(dn_ds)):
                #jan_cog_pos_count[cog_number] += 1
        
        if "product" in gene_annotation[gene]:
            product = gene_annotation[gene]["product"]
            
            if float(dn_ds) > 1:
                jan_product_pos_count[product] += 1

            #elif not np.isfinite(float(dn_ds)):
                #jan_product_pos_count[product] += 1
    
    
    results_avg = defaultdict()
    for cog in jan_avg_cog:
        results_avg[cog] = np.nansum([float(x) for x in jan_avg_cog[cog] if np.isfinite(float(x))]) \
                        / len(jan_avg_cog[cog])
        
    genome_avg_cog["jan"] = results_avg
    
    genome_pos_count["jan"] = jan_cog_pos_count
    
    genome_prod_count["jan"] = jan_product_pos_count
    
    #August
    aug_cog_pos_count = defaultdict(int)
    aug_avg_cog = defaultdict(list)
    aug_product_pos_count = defaultdict(int)
    
    for gene in pN_pS_Aug[genome]:
        dn, ds, dn_ds = pN_pS_Aug[genome][gene]
        
        
        if np.isnan(float(dn_ds)):
            continue

        if "product" in gene_annotation[gene]:
            product = gene_annotation[gene]["product"]
            if float(dn_ds) > 1:                
                product_to_print.append(["August",genome,gene,product,dn,ds,dn_ds])
            if np.isinf(float(dn_ds)):
                product_to_print.append(["August",genome,gene,product,dn,ds,dn_ds])
        
        if np.isinf(float(dn_ds)):
            continue
                
        if "COG" in gene_annotation[gene]:
            cog_number = gene_annotation[gene]["COG"]
        
            aug_avg_cog[cog_number].append(dn_ds)
            
            if float(dn_ds) > 1:
                aug_cog_pos_count[cog_number] += 1
                
                cog_letters = one_letter_code_cog[cog_number]
                
                for entry in cog_letters:
                    cog_gene_description_positive_selection[genome].append(["august",genome,gene,cog_number,cog_letter_desc[entry],str(dn_ds)])
            #elif not np.isfinite(float(dn_ds)):
                #aug_cog_pos_count[cog_number] += 1
                
        if "product" in gene_annotation[gene]:
            product = gene_annotation[gene]["product"]
            
            if float(dn_ds) > 1:
                aug_product_pos_count[product] += 1
            #elif not np.isfinite(float(dn_ds)):
                #aug_product_pos_count[product] += 1
    
    
    results_avg = defaultdict()
    for cog in aug_avg_cog:
        results_avg[cog] = np.nansum([float(x) for x in aug_avg_cog[cog] if np.isfinite(float(x))]) \
                        / len(aug_avg_cog[cog])
        
    genome_avg_cog["aug"] = results_avg
    genome_pos_count["aug"] = aug_cog_pos_count 
    genome_prod_count["aug"] = aug_product_pos_count
    
    average_pn_ps_cogs[genome] = genome_avg_cog
    cog_count_positive_selection[genome] = genome_pos_count
    product_count_positive_selection[genome] = genome_prod_count
```
