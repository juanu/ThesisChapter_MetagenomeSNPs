```python
#Read the file that correlates genome with scaffold
from collections import defaultdict
genome_scaf_info = defaultdict(list)

genomes_order = ["J07HWQ1", "J07HWQ2", "J07HQX50", "J07AB56", "J07AB43", "J07HN4", "J07HN6", "J07HX64",
                 "J07HX5", "J07HB67", "J07HR59", "A07HB70", "A07HR67", "A07HN63", "A07HR60", "J07SB", "G22"]

for line in open("genome_scaffold_list.txt"):
    if line.strip():
        line=line.rstrip()
        genome,scaffold = line.split("\t")
        genome_scaf_info[genome].append(scaffold)

scaffold_length = defaultdict(int)

for line in open("scaffold_length.txt"):
    if line.strip():
        line=line.rstrip()
        scaffold,length = line.split("\t")
        scaffold_length[scaffold] = length
        
genome_length = defaultdict(int)

for line in open("genome_lengths.txt"):
    if line.strip():
        line=line.rstrip()
        genome,length=line.split("\t")
        genome_length[genome]=length
```

```python
gene_coverage_results = "/home/ubuntu/LT_data/New_Reads/Mapping/gene_coverage/"
```

```python
#Process all the genomes
#gene_coverage_genome = gene_coverage_results + "J07HB67.bedtools_annotated"
gene_product = defaultdict()
genome_counts = defaultdict(list) # Format is genome: (gene, length, jan23,jan25,aug07,aug09)
read_count_genomes = defaultdict() # format is key:genome, then jan23, jan 25, aug 07, aug09
gene_coords = defaultdict(list)

for genome in genomes_order:
    gene_coverage_genome = gene_coverage_results + genome + ".bedtools_annotated"
    total_counts = [0,0,0,0] # format is jan23, jan 25, aug 07, aug09

    for line in open(gene_coverage_genome):
        if line.strip():
            diff_jan_aug = 0
            line=line.rstrip()
        
            elements = line.split("\t")
        
            scaffold = elements[0]
            feat_type = elements[2]
            start = int(elements[3])
            end = int(elements[4])
            strand = elements[6]
            info = elements[8]
        
            jan23_count,jan25_count,aug07_count,aug09_count = elements[9:]
        
            feat_length = end - start + 1
            
            id_info = None
            product = None
            
            
            
            for entry in info.split(";"):
                if entry.startswith("ID"):
                    id_info = entry.split("=")[1]
                if entry.startswith("product"):
                    product = entry.split("=")[1]
                    
            gene_product[id_info] = product
            gene_coords[id_info] = [start,end,feat_length]
            
            genome_counts[scaffold].append([id_info, feat_length, jan23_count, jan25_count, aug07_count, aug09_count])
        
            total_counts[0] += int(jan23_count)
            total_counts[1] += int(jan25_count)
            total_counts[2] += int(aug07_count)
            total_counts[3] += int(aug09_count)
    
    read_count_genomes[genome] = total_counts

            
 
        
```

```python
#I'm using a coverage per gene. I need to calculate rpkm for each gene
# For this I need to know the amount of mapped reads in each sample
from fisher import pvalue
reads_sample = {"Jan23": 34900975 , "Jan25":31635270, "Aug07":22109026, "Aug09":18363540}

#Calculate rpkms and perform fisher test        
genome_rpkm = defaultdict(list) # Format is genome: (gene, jan23, jan25, aug07, aug09, )
genome_pvalues = defaultdict(list)

for scaffold in genome_counts:
    for entry in genome_counts[scaffold]:
        gene, length, reads_jan23,reads_jan25,reads_aug07,reads_aug09 = entry
        
        rpkm_jan23 = 1000 * (float(reads_jan23) / int(length)) * (1000000/float(reads_sample["Jan23"]))
        rpkm_jan25 = 1000 * (float(reads_jan25) / int(length)) * (1000000/float(reads_sample["Jan25"]))
        
        rpkm_aug07 = 1000 * (float(reads_aug07) / int(length)) * (1000000/float(reads_sample["Aug07"]))
        rpkm_aug09 = 1000 * (float(reads_aug09) / int(length)) * (1000000/float(reads_sample["Aug09"]))
        
        rpkm_january = 1000 * (float(reads_jan23+reads_jan25) / int(feat_length)) * \
                       (1000000/(float(reads_sample["Jan23"]+reads_sample["Jan25"])))
            
        rpkm_august = 1000 * (float(reads_aug07+reads_aug09) / int(feat_length)) * \
                       (1000000/(float(reads_sample["Aug07"]+reads_sample["Aug09"])))
            

        genome_rpkm[scaffold].append([gene,length,rpkm_jan23,rpkm_jan25,rpkm_aug07,rpkm_aug09,rpkm_january,rpkm_august])
        
        #Calculate fisher for january 23 versus january 25
        p_jan23_jan25 = pvalue(int(reads_jan23),int(reads_jan25),(total_counts[0]-int(reads_jan23)),(total_counts[1]-int(reads_jan25)))
        p_aug07_aug09 = pvalue(int(reads_aug07),int(reads_aug09),(total_counts[2]-int(reads_aug07)),(total_counts[3]-int(reads_aug09)))
        
        p_jan_aug = pvalue(int(reads_jan23)+int(reads_jan25), int(reads_aug07)+int(reads_aug09),
                           ((total_counts[0]+total_counts[1])-(int(reads_jan23)+int(reads_jan25))),
                           ((total_counts[2]+total_counts[3])-(int(reads_aug07)+int(reads_aug09))))
        
        genome_pvalues[scaffold].append([gene,float(p_jan23_jan25.two_tail),float(p_aug07_aug09.two_tail),float(p_jan_aug.two_tail)])
```

```python
#Save the output files, just in case
import cPickle as pickle

pickle.dump(genome_pvalues, open("pvalues_scaffolds_results.txt", 'wb'))
pickle.dump(genome_rpkm, open("scaffolds_rpkm_results.txt", 'wb'))

```

```python
#Print results in a table, for each genome. Only genes that have a significant pvalue

for genome in genomes_order:
 
    pvalue_results_file = open("diff_coverage/" + genome + "_pvalue_results.txt",'w')
    
    for scaffold in genome_scaf_info[genome]:
        
        for entry in genome_pvalues[scaffold]:
            gene, p_jan, p_aug, p_both = entry
            
            product = gene_product[gene]
            
            if 0.05 > p_both:
                #Get the rpkm values. Probably not very efficient, but works right now
                january = 0
                august = 0
                
                for rpkm_values in genome_rpkm[scaffold]:
                    if rpkm_values[0] == gene:
                        january = rpkm_values[6]
                        august = rpkm_values[7]
                        break
                
                month = None
                
                if january > august:
                    month = "january"
                if august > january:
                    month = "august"
                         
                pvalue_results_file.write(scaffold + "\t" + gene + "\t" + str(product) + 
                                          "\t"+ str(january) + "\t" + str(august) + 
                                          "\t" + str(p_both) + "\t" + month + "\n")
                    
    pvalue_results_file.close()
        



```
