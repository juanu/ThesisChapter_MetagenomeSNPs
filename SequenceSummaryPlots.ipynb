```python
import numpy as np
import matplotlib as mpl
#mpl.use('Agg')
from matplotlib import pyplot as plt
import seaborn as sns
from math import modf
```

```python
def calculate_gc_pct(input):
    input = input.lower()
    g_plus_c = input.count('g') + input.count('c')
    gc_pct = (g_plus_c / float(len(input))) * 100
    
    return gc_pct

def get_gc_bin(value):
    b, a = modf(value)
    if b==0:
        return a
    else:
        return a+1
        
```

```python
def summary_gc_results(input_file):
    gc_dict = {x:0 for x in range(0,101)}
    
    for line in open(input_file):
        if line.strip():
            line = line.rstrip()
            length,gc = line.split("\t")
            
            b,a = modf(float(gc))
            
            if b==0:
                gc_dict[a] += int(length)
            
            else:
                gc_dict[a+1] += int(length)
                    
    return gc_dict
            
            
            
            
```

```python
##Stacked plot of the mapping results
from collections import OrderedDict
mapping_results = OrderedDict()

for line in open("MappingAbundances.txt"):
    if line.strip():
        line = line.rstrip()
        genome,p1,p2,p3,p4 = line.split("\t")
        mapping_results[genome] = [int(p1),int(p2),int(p3),int(p4)]
```

```python
#Transform in pct. Order Jan23, Jan25, Aug07, Aug09
total_reads_with_unmapped = []
total_reads_no_unmapped = []

i = 0
while 4 > i:
    total_reads_with_unmapped.append(sum(x[i] for x in mapping_results.values()))
    total_reads_no_unmapped.append(sum(value[i] for key,value in mapping_results.items() if not key=="Unmapped"))
    i += 1


pct_with_unmapped = []
pct_no_unmapped = []

for genome in mapping_results:
    new_entry = []
    i=0
    for value in mapping_results[genome]:
        new_entry.append(float(value)/total_reads_with_unmapped[i]*100)
        i += 1
    
    pct_with_unmapped.append(new_entry)

for genome in mapping_results:
    if genome=="Unmapped":
        continue
    else:
        new_entry = []
        i = 0
        for value in mapping_results[genome]:
            new_entry.append(float(value)/total_reads_no_unmapped[i]*100)
            i += 1
        pct_no_unmapped.append(new_entry)
```

```python
#np.array(mapping_results.values())
fig = plt.figure(figsize=(12,6))
plt.stackplot(np.arange(4),np.array(pct_with_unmapped))
plt.xlim(0,3)
plt.ylim(0,100)
```

```python
#np.array(mapping_results.values())
fig = plt.figure(figsize=(12,6))
plt.stackplot(np.arange(4),np.array(pct_no_unmapped))
plt.xlim(0,3)
plt.ylim(0,100)
```
