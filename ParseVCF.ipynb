```python
#Summary of each genome
snp_eff_folder = "/home/ubuntu/LT_data/New_Reads/Mapping/variation/separated_results/snpEff_results/"
results_snp_types = defaultdict()

for genome in genomes_order:
    genome_results = defaultdict()
    for sample in sample_ids:
        sample_results = defaultdict(int)
        
        snpEff_results = snp_eff_folder + "snpEff_" + genome + "_" + sample + ".vcf"
        
        vcf_reader = vcf.Reader(open(snpEff_results,'r'))
        
        for record in vcf_reader:
            if not record.INFO.has_key("EFF"):
                continue
            if record.INFO["EFF"][0].startswith("SYNONYMOUS_CODING"):
                sample_results["synonymous"] += 1
            if record.INFO["EFF"][0].startswith("NON_SYNONYMOUS_CODING"):
                sample_results["nonsynon"] += 1
            if record.INFO["EFF"][0].startswith("INTERGENIC"):
                sample_results["intergenic"] += 1
        
        genome_results[sample] = sample_results
    
    results_snp_types[genome] = genome_results
         
```

```python
fig, axes = plt.subplots(nrows=6, ncols=3,figsize=(10,15))
i = 0
j = 0

plot_types_boxplot = defaultdict(list)


for genome in genomes_order:

    plot_dict = defaultdict(list)
    
    for sample in sample_ids:
        for snp_type in results_snp_types[genome][sample]:
            plot_dict[snp_type].append(results_snp_types[genome][sample][snp_type])
            
    pct_plot_dict = defaultdict(list)
    
    for entry in plot_dict:
        m = 0
        for value in plot_dict[entry]:
            pct_value = int(value) / float(sum([x[m] for x in plot_dict.values()])) * 100
                                           
            pct_plot_dict[entry].append(pct_value)
            
            m += 1
    
    #print genome, "Intergenic", sum(pct_plot_dict["intergenic"])/len(pct_plot_dict["intergenic"])
    #print genome, "Synonymous", sum(pct_plot_dict["synonymous"])/len(pct_plot_dict["synonymous"])
    #print genome, "NonSyno", sum(pct_plot_dict["nonsynon"])/len(pct_plot_dict["nonsynon"])
    #print ""
    
    plot_types_boxplot["Intergenic"].append(sum(pct_plot_dict["intergenic"])/len(pct_plot_dict["intergenic"]))
    plot_types_boxplot["Synonymous"].append(sum(pct_plot_dict["synonymous"])/len(pct_plot_dict["synonymous"]))
    plot_types_boxplot["NonSyno"].append(sum(pct_plot_dict["nonsynon"])/len(pct_plot_dict["nonsynon"]))
    
    df = pd.DataFrame(pct_plot_dict, index=sample_ids)
    df.plot(ax=axes[j,i],kind="bar",stacked=True,legend=False,title=genome,fontsize=18)
    #df.plot(

    
    if i == 0:
        i += 1
    elif i == 1:
        i += 1
    elif i == 2:
        i = 0
        j += 1

patches, labels = axes[1,1].get_legend_handles_labels()

plt.figlegend(patches, ["Intergenic", "Non-Synonymous", "Synonymous"], loc = "lower right",bbox_to_anchor= (-0.01,0.07,1,1), bbox_transform=plt.gcf().transFigure)
        
fig.delaxes(axes[5,2])
plt.tight_layout()
plt.savefig("SNPsummary.pdf")
```
