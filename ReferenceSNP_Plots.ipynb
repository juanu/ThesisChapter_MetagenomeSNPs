```python
#Import the modules needed
from collections import defaultdict
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
from collections import defaultdict
import re
```

```python
N = 11 # Number of genomes

#The plot nonsynon, syn, intergenic/rest

#Format intergenic, non synon, synon
J07HN4 = [11.821, 18.646, 69.031]
J07HN6 = [10.262, 20.219, 69.121]
J07HX64 = [15.596, 27.198, 56.785]
J07HB67 = [6.871, 25.269, 67.372]
J07HQX50 = [25, 33.75, 41.25]
J07HR59 = [13.758, 25.168, 60.067]
J07HQW1 = [37.886, 22.402, 39.019]
J07HQW2 = [36.05, 26.407, 36.818]
J07HX5 = [10.932, 21.571, 67.252]
#J07AB43 has a lot of frame_shifts 48.203
#exon: 91.278
J07AB43 = [8.705, 20.524, 19.625]
#J07AB56, 0.163 frameshifts
J07AB56 = [5.809, 19.185, 74.366]
```

```python
#Make stacked barplot
genomes_names = ["J07HQW1", "J07HQW2", "J07HQX50", "J07AB56", "J07AB43",
           "J07HN4", "J07HN6", "J07HX64", "J07HX5", "J07HB67", "J07HR59"]

genome_list = [J07HQW1, J07HQW2, J07HQX50, J07AB56, J07AB43, J07HN4,
               J07HN6, J07HX64, J07HX5, J07HB67, J07HR59]

width = 0.35

#Function to add the values
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),ha='center',
                va='bottom')


fig = plt.figure(figsize=(14,6))

plot_intergen = plt.bar(np.arange(N), [x[0] for x in genome_list], width, color='#66c2a5')
plot_non_syn = plt.bar(np.arange(N), [x[1] for x in genome_list], width, color='#fc8d62', bottom=[x[0] for x in genome_list])
plot_synon= plt.bar(np.arange(N), [x[2] for x in genome_list], width, color='#8da0cb', bottom=[x[1]+x[0] for x in genome_list])
plot_other = plt.bar(np.arange(N), [100-sum(x) for x in genome_list],width, 
                     color='#e78ac3', bottom=[sum(x) for x in genome_list])



plt.ylabel("Type of SNP (%)",fontsize=16)
#plt.title("Percentage of SNPs by type",fontsize=17)
plt.xticks(np.arange(N) + width/2., (genomes_names), rotation=45,fontsize=18)
plt.yticks(np.arange(0,100,10),fontsize=15)

#Adjust legend
plt.legend((plot_intergen[0],plot_non_syn[0],plot_synon[0],plot_other[0]),
           ("Intergenic", "Non synonymous", "Synonymous", "Other"),fontsize=15,loc=1)

#Add values to each bar
for bar,value in zip(plot_intergen,genome_list):
    height = value[0] / 2. - 1
    plt.text(bar.get_x() + bar.get_width()/2. , height, '%d'%int(value[0]), ha='center', va='bottom', fontsize=13)

for bar,value in zip(plot_non_syn,genome_list):
    height = value[0] + value[1] / 2. - 1
    plt.text(bar.get_x() + bar.get_width()/2. , height, '%d'%int(value[1]), ha='center', va='bottom', fontsize=13)

for bar,value in zip(plot_synon,genome_list):
    height = value[0] + value[1] + value[2] / 2. - 1
    plt.text(bar.get_x() + bar.get_width()/2. , height, '%d'%int(value[2]), ha='center', va='bottom', fontsize=13)

#J07AB43 has value in other. Add manually
J07AB43_x = plot_other[4].get_x() + plot_other[4].get_width()/2.
J07AB43_y = sum(genome_list[4]) + (100-sum(genome_list[4])) / 2.
plt.text(J07AB43_x, J07AB43_y, '%d'%int(100-sum(genome_list[4])), ha='center', va='bottom', fontsize=13)



plt.tight_layout()
plt.savefig("/Users/juan/Documents/Thesis/Thesis_LaTex/Chapter5/Figures/ReferenceSNPtypes.pdf")


```
