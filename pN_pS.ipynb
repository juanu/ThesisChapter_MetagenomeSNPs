```python
def calculate_dn_ds(input_gff,input_vcf,input_fasta):
    import re
    from Bio import SeqIO
    from Bio.Seq import Seq
    from cogent import LoadSeqs,DNA
    
    #Read the GFF
    gene_info = defaultdict(lambda:defaultdict())
    for line in open(input_gff):
        if line.strip():
            if line.startswith("#"):
                continue
            
            line = line.rstrip()
            elements = line.split("\t")
            #save type, start, stop, orientation
            gene_info[elements[0]][elements[8].split(";")[0].split("=")[1]]=[elements[2],elements[3],elements[4],elements[6]]
    
    #Read the VCF
    snp_gene_info = defaultdict(lambda:defaultdict())
    
    vcf_reader = vcf.Reader(open(input_vcf, 'r'))
    
    for record in vcf_reader:
        if record.INFO["TYPE"][0] == "snp":
            snp_gene_info[record.CHROM][record.POS] = (str(record.REF), str(record.ALT[0]))
    
    #Read the fasta file
    record_dict = SeqIO.index(input_fasta, "fasta")
    
    #Generate snps by gene
    snps_by_gene = defaultdict(lambda:defaultdict(list))
    
    for scaffold in gene_info:
        for gene in gene_info[scaffold]:
            
            start = int(gene_info[scaffold][gene][1])
            stop = int(gene_info[scaffold][gene][2])
            
            for position in range(start,stop+1):
                if position in snp_gene_info[scaffold].keys():
                    
                    #Only store coding sequences
                    if not gene_info[scaffold][gene][0] == "CDS":
                        continue

                    position_in_gene = int(position) - start + 1
                    snps_by_gene[gene][position_in_gene] = snp_gene_info[scaffold][position], position
            
    #########
    #Calculate pN/pS
    mixture_dict={'N':'ATGC'}
    mixture_re = re.compile("[N]")
    stop_codons=["TAA","TGA","TAG"]
    gene_dn_ds_list=[]
    
    #Get the fasta sequence per gene
    fasta_sequence_gene = defaultdict(str)
    for scaffold in gene_info:
        scaf_sequence = record_dict[scaffold].seq
        
        for gene in gene_info[scaffold]:
            feat_type, start, stop, strand = gene_info[scaffold][gene]
            
            if feat_type == "CDS":
                
                dna_sequence = str(scaf_sequence[int(start)-1:int(stop)])
                
                if strand == "-":
                    fasta_sequence_gene[gene] = str(Seq(dna_sequence).reverse_complement()) # Use Biopython to reverse complement
                else:
                    fasta_sequence_gene[gene] = str(scaf_sequence[int(start)-1:int(stop)])
                
    
    #Calculate pn_ps
    gene_pn_ps = defaultdict(list)
    
    for gene in snps_by_gene:
        if len(snps_by_gene[gene]) == 0:
            continue
        
        #Get sequence and codon list
        seq_object_pycogent = DNA.makeSequence(fasta_sequence_gene[gene], gene)
        codon_list = seq_object_pycogent.getInMotifSize(3)
        
        #Calculate expected NS and S values
        gene_expected_NS = 0
        gene_expected_S = 0
        gene_observed_NS = 0
        gene_observed_S = 0
        
        for codon in codon_list:
            #Check for stop codons
            if codon in stop_codons:
                continue
            
            #Deal with Ns
            mixtures = mixture_re.findall(codon)
            
            if len(mixtures) > 1:
                #codon contains more than one mixture, no way to resolve
                continue
            
            elif len(mixtures)==1:
                resolutions = mixture_dict[mixtures[0]]
                means = [0,0]
                
                for r in resolutions:
                    rcodon = codon.replace(mixtures[0],r)
                    
                    if rcodon in stop_codons:
                        break
                    else:
                        position = 0
                        
                        while position < 3:
                            index = position*2
                            means[0] += float(codon_site_dict[rcodon][index] / float(len(resolutions)))
                            means[1] += float(codon_site_dict[rcodon][index+1]) / float(len(resolutions))
                            position += 1
                
                
                gene_expected_NS =+ means[0]
                gene_expected_S =+ means[1]
                continue                
        
            #Iterate over each position of the codon
            position = 0
            while position < 3:
                index = position*2
                expected_NS,expected_S = tuple(codon_site_dict[codon][index:index+2])
                position += 1
                gene_expected_NS += expected_NS
                gene_expected_S += expected_S  

        #Calculate the observed frequencies  
    
        codon_errors = 0
        errors = []
        
        for snp in snps_by_gene[gene]:
            codon_division = 0
            codon_number = 0
            snp_in_codon_position = 0
            codon_in_list = 0
            take_complementary_base = 0
            codon_division = snp%3
            codon_number = snp/3
            
            if codon_division==1:
                snp_in_codon_position=1
                codon_number==codon_number+1
            if codon_division==2:
                snp_in_codon_position=2
                codon_number=codon_number+1
            if codon_division==0:
                snp_in_codon_position=3
                
            #Get the codon information
            try:
                consensus_codon=codon_list[codon_number-1]
            except IndexError:
                codon_errors += 1
    
            bases_codon=list(consensus_codon)
                                    
            if consensus_codon in codon_dict:
                consensus_aminoacid=codon_dict[consensus_codon]
            else:
                #Skips N and failed results
                errors.append([gene,consensus_codon])
                continue
                        
            new_codon = []
            try:
                ref_base,new_base = snps_by_gene[gene][snp][0]
            except ValueError:
                continue
                            
            
            if snp_in_codon_position==1:
                new_codon=str(new_base),bases_codon[1],bases_codon[2]
            elif snp_in_codon_position==2:
                new_codon=bases_codon[0],str(new_base),bases_codon[2]
            elif snp_in_codon_position==3:
                new_codon=bases_codon[0],bases_codon[1],str(new_base)
                              
            try:
                new_aminoacid = codon_dict[''.join(new_codon)]
            except KeyError:
                continue
            
            #Add into observed NS or S
            
            if new_aminoacid==consensus_aminoacid:
                gene_observed_S += 1
            else:
                gene_observed_NS += 1
    
        #Calculate dn/ds ratio
    
        if gene_expected_NS > 0:
            dn=float(gene_observed_NS) / gene_expected_NS
        else:
            dn = 0
            
        if gene_expected_S > 0:
            ds=float(gene_observed_S) / gene_expected_S
        else:
            ds = 0
        
        if ds>0:
            dn_ds=round(dn/ds,3)
        else:
            if dn == 0:
                dn_ds = "NaN"
            else:
                dn_ds = "Inf"
        
        gene_pn_ps[gene] = [dn,ds,dn_ds]
        
    return gene_pn_ps
    
        
    
```
