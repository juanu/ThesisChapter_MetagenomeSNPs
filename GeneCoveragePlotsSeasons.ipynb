```python
#Read annotations
annotation_folder = "/home/ubuntu/LT_data/CompGenomics/LT_Genomes/v2/processed_genomes/annotation/"
file_letter_code_cog = "list_cog"
```

```python
#Read annotation and store
cog_annotation = defaultdict()
for genome in genomes_order:
    for line in open(annotation_folder + genome + ".txt"):
        if line.strip():
            line = line.rstrip()
            elements = line.split("\t")
            if elements[1] == "COG":
                cog_annotation[elements[0]] = (elements[2],elements[3])
                
        
```

```python
one_letter_code_cog = defaultdict(list)
for line in open("list_cog", 'r'):
    if line.strip():
        line = line.rstrip()
        letter, number, description = line.split("\t")
        
        for char in letter:
            if char=="[" or char=="]":
                continue
            else:
                one_letter_code_cog[number].append(char)
```

```python
#Genes with diff coverage
diff_coverage_folder = "/home/ubuntu/Notebooks/diff_coverage/"
```

```python
cog_count_genome = defaultdict()

cog_count_letter_genome = defaultdict()

for genome in genomes_order:
    cog_january = defaultdict(int)
    cog_august = defaultdict(int)
    cog_summary = defaultdict()
    
    letter_count_january = defaultdict(int)
    letter_count_august = defaultdict(int)
    letter_count_summary = defaultdict()
    
    for line in open(diff_coverage_folder + genome + "_pvalue_results.txt",'r'):
        if line.strip():
            line = line.rstrip()
            elements = line.split("\t")
            gene_id = elements[1]
            product = elements[2]
            month = elements[6]
            
            cog = (None, None)
            if gene_id in cog_annotation:
                cog = cog_annotation[gene_id]
                            
            if month == "january":
                cog_january[cog[0]] += 1
                
                if cog[0] in one_letter_code_cog:
                    for letter in one_letter_code_cog[cog[0]]:
                        letter_count_january[letter] += 1
                        
                else:
                    letter_count_january[None] += 1
                
            if month == "august":
                cog_august[cog[0]] += 1
                
                if cog[0] in one_letter_code_cog:
                    for letter in one_letter_code_cog[cog[0]]:
                        letter_count_august[letter] += 1
                
                else:
                    letter_count_august[None] += 1
                
    
    cog_summary["january"] = cog_january
    cog_summary["august"] = cog_august
    
    cog_count_genome[genome] = cog_summary
    
    letter_count_summary["january"] = letter_count_january
    letter_count_summary["august"] = letter_count_august
    
    cog_count_letter_genome[genome] = letter_count_summary
            
        
            
```
