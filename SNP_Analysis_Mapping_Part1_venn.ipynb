```python
import vcf
from collections import defaultdict
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import pybedtools
from matplotlib_venn import venn2
```

```python
#All the genome scaffold information
from collections import defaultdict
genome_scaf_info = defaultdict(list)

genomes_order = ["J07HWQ1", "J07HWQ2", "J07HQX50", "J07AB56", "J07AB43", "J07HN4", "J07HN6", "J07HX64",
                 "J07HX5", "J07HB67", "J07HR59", "A07HB70", "A07HR67", "A07HN63", "A07HR60", "J07SB", "G22"]

sample_ids = ["Jan23", "Jan25", "Aug07", "Aug09"]

for line in open("genome_scaffold_list.txt"):
    if line.strip():
        line=line.rstrip()
        genome,scaffold = line.split("\t")
        genome_scaf_info[genome].append(scaffold)

scaffold_length = defaultdict(int)

for line in open("scaffold_length.txt"):
    if line.strip():
        line=line.rstrip()
        scaffold,length = line.split("\t")
        scaffold_length[scaffold] = length
        
genome_length = defaultdict(int)

for line in open("genome_lengths.txt"):
    if line.strip():
        line=line.rstrip()
        genome,length=line.split("\t")
        genome_length[genome]=length
```

```python
#Comparison by sample type

jan_comp = defaultdict(lambda:defaultdict())
aug_comp = defaultdict(lambda:defaultdict())
for genome in genomes_order:
    
    jan23 = pybedtools.BedTool(vcf_sample_files + genome + "_Jan23.vcf")
    jan25 = pybedtools.BedTool(vcf_sample_files + genome + "_Jan25.vcf")
    aug07 = pybedtools.BedTool(vcf_sample_files + genome + "_Aug07.vcf")
    aug09 = pybedtools.BedTool(vcf_sample_files + genome + "_Aug09.vcf")
    
    #january
    jan_comp[genome]["jan23"] = (jan23-jan25).count()
    jan_comp[genome]["jan25"] = (jan25-jan23).count()
    jan_comp[genome]["both"] = (jan23+jan25).count()
    
    #august
    aug_comp[genome]["aug07"] = (aug07-aug09).count()
    aug_comp[genome]["aug09"] = (aug09-aug07).count()
    aug_comp[genome]["both"] = (aug07+aug09).count()
```

```python
#Path to VCF files per month:
vcf_month_file = "/home/ubuntu/LT_data/New_Reads/Mapping/variation/separated_results/by_month/"
```

```python
#Comparison by month 

season_comp = defaultdict(lambda:defaultdict())


for genome in genomes_order:
    
    january = pybedtools.BedTool(vcf_month_file + genome + "_Jan.vcf")
    august = pybedtools.BedTool(vcf_month_file + genome + "_Aug.vcf")

    #Comparison
    season_comp[genome]["jan"] = (january-august).count()
    season_comp[genome]["aug"] = (august-january).count()
    season_comp[genome]["both"] = (january+august).count()

```
